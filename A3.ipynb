```python
data['Gender'] = data['Gender'].map({'Female' : 0, 'Male' : 1})
data.describe()
data.std()
```

```python
data['Cohort'] = data['Cohort'].replace('Under 16', '0-16')
```

```python
data['Cohort'] = data['Cohort'].replace('16', '0-16')
data['Cohort'] = data['Cohort'].replace('17', '16-17')
data['Cohort'] = data['Cohort'].replace('18', '17-18')
data['Cohort'] = data['Cohort'].replace('19', '18-19')
data['Cohort'] = data['Cohort'].replace('20', '19-20')
data['Cohort'] = data['Cohort'].replace('21', '20-21')
data['Cohort'] = data['Cohort'].replace('22', '21-22')
data['Cohort'] = data['Cohort'].replace('23', '22-23')
data['Cohort'] = data['Cohort'].replace('24', '23-24')
data['Cohort'] = data['Cohort'].replace('85', '85-100')
```

```python
numeric_columns = data.select_dtypes(include=[np.number]).columns
numeric_columns = numeric_columns[numeric_columns != 'State']  # Exclude 'State' column
std_excluding_state = data[numeric_columns].std()
std_excluding_state
```
